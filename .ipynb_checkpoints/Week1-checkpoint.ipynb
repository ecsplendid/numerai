```python
train_x = data.x[ data.region=='train' ]
val_x = data.x[ data.region=='validation' ]
train_y = data.y_df['bernie'][ data.region=='train' ].values
val_y = data.y_df['bernie'][ data.region=='validation' ].values
```

```python
import matplotlib.pyplot as plt

def plot(history, val=True, train=True, title='Bernie'):
    if train:
        plt.plot(history.history['loss'][2:])
    if val:
        plt.plot(history.history['val_loss'][2:])
        
    plt.title('model loss on %s!' %(title))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    
    labels = []
    if train: 
        labels.append('train')
    if val: 
        labels.append('validation')
        
    plt.legend(labels, loc='upper left')
    plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras import regularizers

model_dense = Sequential([
    Dense(32, input_shape=(50,)),
    Activation('relu'),
    Dropout(0.3),
    Dense(100),
    Activation('relu'),
    Dropout(0.3),
    Dense(100),
    Activation('relu'),
    Dropout(0.3),
    Dense(100),
    Activation('relu'),
    Dropout(0.3),
    Dense(1),
    Activation('sigmoid'),
])


model_dense.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['binary_crossentropy'])

history = model_dense.fit(train_x, 
                    train_y, 
                    validation_data=(val_x,val_y), 
                    epochs=200, 
                    batch_size=5000, 
                    shuffle=False)

plot(history)
```

```python
from more_itertools import *
from pipe import *
import math
```

```python
@Pipe
def as_npy(l):
    return np.array(list(l))
```

```python
def getX(signals,step=2,window_size=10): 
    x = windowed( range(signals.shape[0]), window_size, step=step) \
    | select( lambda x: x | select(lambda y: signals[y]) | as_npy ) \
    | take((signals.shape[0]/step)-window_size) \
    | as_npy
    return x
```

```python
def getY(labels,step=2,window_size=10):
    return labels[window_size:math.ceil((((labels.shape[0]/step)-window_size)+window_size))]
```

```python
args = {"step":1, "window_size":5}

X = getX(train_x, **args)
Y = getY(train_y, **args)
X_val = getX(val_x, **args)
Y_val = getY(val_y, **args)
```

```python
from keras.models import *
from keras.layers import *
from keras import regularizers

# cnn idea
cnn_model = Sequential([
    Conv1D(10, kernel_size=3,input_shape=(args['window_size'],50)),
    Activation('relu'),
    SpatialDropout1D(0.2),
    Conv1D(10, kernel_size=3,input_shape=(args['window_size'],50)),
    Activation('relu'),
    SpatialDropout1D(0.2),
    Flatten(),
    Dropout(0.3),
    Dense(100),
    Dropout(0.3),   
    Dense(1),
    Activation('sigmoid'),
])

model.summary()

model.compile(optimizer='adam',
              loss='binary_crossentropy')

```

```python
history = model.fit(X, 
                    Y, 
                    epochs=50, 
                    batch_size=2000, 
                    shuffle=False,
                    validation_data=(X_val, Y_val))

plot(history)
#0.6925 best
```

```python
from keras.models import *
from keras.layers import *
from keras import regularizers

input_cnn = Input(shape=(args['window_size'],50))
input_dense = Input(shape=(50,))

# cnn idea
c1 = Conv1D(10, kernel_size=3, activation='relu')(input_cnn)
c1 = SpatialDropout1D(0.2)(c1)
c1 = Conv1D(10, kernel_size=3, activation='relu')(c1)
c1 = SpatialDropout1D(0.2)(c1)
c1 = Flatten()(c1)
c1 = Dropout(0.3)(c1)
c1 = Dense(100, activation='relu')(c1)

d1 = Dense(32, activation='relu')(input_dense)
d1 = Dropout(0.3)(d1)
d1 = BatchNormalization()(d1)
d1 = Dense(100, activation='relu')(d1)
d1 = Dropout(0.3)(d1)
d1 = Dense(100, activation='relu')(d1)
d1 = Dropout(0.3)(d1)
d1 = Dense(100, activation='relu')(d1)

merge_1 = concatenate([c1, d1])

merge_1 = Dense(100)(merge_1)
merge_1 = Dropout(0.3)(merge_1)
merge_1 = Dense(70)(merge_1)
merge_1 = Dropout(0.3)(merge_1)
merge_1 = Dense(60)(merge_1)
merge_1 = Dropout(0.3)(merge_1)
merge_1 = Dense(10)(merge_1)
merge_1 = Dense(1)(merge_1)

model = Model(inputs=[input_cnn,input_dense],outputs=merge_1)

model.summary()

model.compile(optimizer='adam',
              loss='binary_crossentropy')
```

```python
def residual_block(y, size=200, drop=0.4):
    shortcut = y

    y = Dense(size)(y)
    y = BatchNormalization()(y)
    y = Dropout(drop)(y)
    y = LeakyReLU()(y)
    y = Dense(size)(y)
    y = BatchNormalization()(y)
    y = Dropout(drop)(y)
    y = LeakyReLU()(y)
    y = Dense(size)(y)
    y = BatchNormalization()(y)
    y = Dropout(drop)(y)
    y = LeakyReLU()(y)
    y = add([shortcut, y])
    y = LeakyReLU()(y)

    return y
```

```python
from keras.models import *
from keras.layers import *
from keras import regularizers

input_dense1 = Input(shape=(50,))
input_dense2 = Input(shape=(50,))
input_dense3 = Input(shape=(50,))

d1 = Dense(50)(input_dense1)
d1 = residual_block(d1,size=50)
d1 = residual_block(d1,size=50)
d1 = residual_block(d1,size=50)
d1 = Dense(100)(d1)

d2 = Dense(300)(input_dense2)
d2 = residual_block(d2,size=300)
d2 = residual_block(d2,size=300)
d2 = residual_block(d2,size=300)
d2 = Dense(100)(d2)

dd = Dense(100)(input_dense3)
dd = residual_block(dd,size=100)
dd = residual_block(dd,size=100)
dd = residual_block(dd,size=100)
dd = Dense(100)(dd)

merge_1 = concatenate([dd, d2, d1])
merge_1 = LeakyReLU()(merge_1)
merge_1 = Dense(200)(merge_1)

d4 = residual_block(merge_1)
d4 = residual_block(d4)
d4 = residual_block(d4)
d4 = residual_block(d4)
d4 = Dense(1, activation='sigmoid')(d4)

model = Model(inputs=[input_dense1,input_dense2,input_dense3],outputs=d4)

model.summary()

model.compile(optimizer='adam',
              loss='binary_crossentropy')
```

```python
history = model.fit([train_x, train_x,  train_x], 
                    train_y, 
                    epochs=50, 
                    batch_size=4000, 
                    shuffle=True,
                    validation_data = ([val_x, val_x, val_x], val_y)
                   )

plot(history, title='elizabeth')
```

```python
from keras.models import *
from keras.layers import *
from keras import regularizers
from keras.callbacks import ModelCheckpoint

input_dense1 = Input(shape=(50,))

d1 = Dense(100)(input_dense1)
d1 = LeakyReLU()(d1)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = residual_block(d1,size=100, drop=0.6)
d1 = Dense(1, activation='sigmoid')(d1)

model = Model(inputs=[input_dense1],outputs=d1)

model.summary()

model.compile(optimizer='adam',
              loss='binary_crossentropy')

checkpoint = ModelCheckpoint(
    'model-{epoch:03d}.h5', 
    verbose=1, 
    monitor='val_loss',
    save_best_only=True, 
    mode='auto')  

history = model.fit([train_x], 
                    train_y, 
                    epochs=50, 
                    batch_size=4000, 
                    shuffle=True,
                    validation_data = ([val_x], val_y),
                    callbacks=[checkpoint]
                   )

plot(history, title='bernie')
```
